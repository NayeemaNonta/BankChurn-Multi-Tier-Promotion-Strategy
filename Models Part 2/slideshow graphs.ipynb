```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 
# Set Seaborn style
sns.set_palette("colorblind")
sns.set_context("notebook")
sns.set_style("whitegrid")


# Data for NN metrics
thresholds_nn = [0.2, 0.5, 0.7]
specificity_nn = [0.8167, 0.9486, 0.9786]
sensitivity_nn = [0.6688, 0.4282, 0.2481]
accuracy_nn = [78.79, 84.71, 83.61]
accuracy_nn_decimal = [accuracy / 100 for accuracy in accuracy_nn]  # Convert accuracy to decimal
precision_nn = [0.4694, 0.6688, 0.7375]
recall_nn = [0.6688, 0.4282, 0.2481]

# Plotting
fig, ax = plt.subplots(figsize=(12, 8), dpi=200)

# Setting the positions for the bars
bar_width = 0.15
index = np.arange(len(thresholds_nn))

# Plotting NN metrics bars
bars_specificity_nn = ax.bar(index - 2*bar_width, specificity_nn, bar_width, label='Specificity')
bars_sensitivity_nn = ax.bar(index - bar_width, sensitivity_nn, bar_width, label='Sensitivity')
bars_accuracy_nn = ax.bar(index, accuracy_nn_decimal, bar_width, label='Accuracy')
bars_precision_nn = ax.bar(index + bar_width, precision_nn, bar_width, label='Precision')
bars_recall_nn = ax.bar(index + 2*bar_width, recall_nn, bar_width, label='Recall')

# Adding labels, title, and legend for NN
ax.set_xlabel('Threshold')
ax.set_ylabel('Metrics')
ax.set_title('Comparison of NN Metrics for Different Thresholds', fontsize=18)
ax.set_xticks(index)
ax.set_xticklabels(thresholds_nn)
ax.legend(loc='upper right')

# Adding annotations for NN
def annotate_bars(bars, values):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.annotate('{:.3f}'.format(value),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
annotate_bars(bars_specificity_nn, specificity_nn)
annotate_bars(bars_sensitivity_nn, sensitivity_nn)
annotate_bars(bars_accuracy_nn, accuracy_nn_decimal)
annotate_bars(bars_precision_nn, precision_nn)
annotate_bars(bars_recall_nn, recall_nn)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

# Set Seaborn style
sns.set_palette("colorblind")
sns.set_context("notebook")

# Data for logistic regression metrics
thresholds_lr = [0.2, 0.5, 0.7]
specificity_lr = [89.88, 81.28, 80.88]
sensitivity_lr = [35.80, 77.97, 81.82]
accuracy_lr = [70.10, 81.24, 80.88]
accuracy_lr_decimal = [accuracy / 100 for accuracy in accuracy_lr]  # Convert accuracy to decimal
precision_lr = [67.1, 5.35, 2.61]

# Convert metrics to decimal form
specificity_lr_decimal = [specificity / 100 for specificity in specificity_lr]
sensitivity_lr_decimal = [sensitivity / 100 for sensitivity in sensitivity_lr]
precision_lr_decimal = [precision / 100 for precision in precision_lr]

# Plotting
fig, ax = plt.subplots(figsize=(10, 8), dpi=200)

# Setting the positions for the bars
bar_width = 0.15
index = np.arange(len(thresholds_lr))

# Plotting logistic regression metrics bars


bars_accuracy_lr = ax.bar(index, accuracy_lr_decimal, bar_width, label='Accuracy')
bars_precision_lr = ax.bar(index + bar_width, precision_lr_decimal, bar_width, label='Precision')
bars_sensitivity_lr = ax.bar(index - bar_width, sensitivity_lr_decimal, bar_width, label='Sensitivity')
bars_specificity_lr = ax.bar(index - 2*bar_width, specificity_lr_decimal, bar_width, label='Specificity')

# Adding labels, title, and legend for logistic regression
ax.set_xlabel('Threshold')
ax.set_ylabel('Metrics')
ax.set_title('Comparison of Logistic Regression Metrics for Different Thresholds', fontsize=18)
ax.set_xticks(index)
ax.set_xticklabels(thresholds_lr)
ax.legend(bbox_to_anchor=(0.9, 1))

# Adding annotations for logistic regression
def annotate_bars(bars, values):
    for bar, value in zip(bars, values):
        height = bar.get_height()
        ax.annotate('{:.2f}'.format(value),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

annotate_bars(bars_sensitivity_lr, sensitivity_lr_decimal)
annotate_bars(bars_accuracy_lr, accuracy_lr_decimal)
annotate_bars(bars_precision_lr, precision_lr_decimal)
annotate_bars(bars_specificity_lr, specificity_lr_decimal)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for logistic regression feature z-values
features = ['vintage', 'age', 'dependents', 'city', 'customer_nw_category', 'branch_code',
            'current_balance', 'previous_month_end_balance', 'average_monthly_balance_prevQ',
            'average_monthly_balance_prevQ2', 'current_month_credit', 'previous_month_credit',
            'current_month_debit', 'previous_month_debit', 'current_month_balance',
            'previous_month_balance', 'days_since_last_transaction', 'gender_Male',
            'occupation_retired', 'occupation_salaried', 'occupation_self_employed',
            'occupation_student']
z_values = [0.080, -3.236, 1.749, 0.555, 2.345, 3.633, -14.898, -2.667, 6.948, -0.676,
            -2.340, -2.242, 2.265, 3.205, 2.389, 0.981, -6.024, 2.511, 0.443, 0.391, 0.644, 0.465]

# Plotting
plt.figure(figsize=(10, 8))
bars = plt.barh(features, z_values, color='skyblue')
plt.xlabel('Z-value')
plt.ylabel('Feature')
plt.title('Logistic Regression Feature Importance Analysis')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to display features from top to bottom

# Add z-values as text annotations
for bar, z_value in zip(bars, z_values):
    plt.text(bar.get_width() * 1.05, bar.get_y() + bar.get_height() / 2, '{:.3f}'.format(z_value),
             va='center', ha='left', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for logistic regression feature p-values and z-values
top_features = ['current_balance', 'days_since_last_transaction', 'average_monthly_balance_prevQ',
                'previous_month_debit', 'age']
p_values = [2e-16, 1.71e-09, 3.69e-12, 0.00135, 0.00121]
z_values = [-14.898, -6.024, 6.948, 3.205, -3.236]

# Plotting
plt.figure(figsize=(14, 6), dpi=200)
bars = plt.barh(range(len(top_features)), z_values, color='skyblue')
plt.yticks(range(len(top_features)), [f'{feature} \n(p={p:.2e})' for feature, p in zip(top_features, p_values)])
plt.xlabel('Z-value')
plt.title('Top 5 Features Importance Analysis (Logistic Regression)', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().invert_yaxis()  # Invert y-axis to display features from top to bottom

# Add z-values as text annotations
for bar, z_value in zip(bars, z_values):
    plt.text(bar.get_width() * 1.05, bar.get_y() + bar.get_height() / 2, '{:.2f}'.format(z_value),
             va='center', ha='left', fontsize=14)

plt.tight_layout()
plt.show()
```
